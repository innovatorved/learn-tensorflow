```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```
# f=c×1.8+32
x = np.arange(-1000 , 1000).astype("float32")
y = np.array([(c*1.8)+32 for c in x]).astype("float32")
```

```
# Dense Layer
l0 = tf.keras.layers.InputLayer(input_shape=[1])  # input layer with shape 
l1 = tf.keras.layers.Dense(units = 1) # output layer units represent the output
```

```
# print(tf.keras.layers.Input.__doc__)
```

```
model = tf.keras.Sequential([l0 , l1])
model.compile(
    loss = "mean_squared_error",
    optimizer=tf.keras.optimizers.Adam(0.1)
)
```

```
model_his = model.fit(x , y , epochs=500, verbose=False)
print("Model fitted Successfully")
```

```
print(f"Get Weights : {l1.get_weights()}")

# loss representation
plt.figure()
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(model_his.history['loss'])
```

```
def predictC2F(n):
    """
    @ param Celcius Value you want to change in to Fahrenheit
    """
    return f"{model.predict([n])[0][0]} F"
```

```
predictC2F(100)
```
